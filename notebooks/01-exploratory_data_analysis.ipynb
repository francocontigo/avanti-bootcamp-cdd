## 1.1 Visualização dos dados

```python
# Relação entre preço e peso (carat)
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=df)
plt.title('Relação entre Peso (Carat) e Preço')
plt.xlabel('Peso (Carat)')
plt.ylabel('Preço')
plt.show()
```

```python
# Qualidade do corte (cut) e preço
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=df)
plt.title('Qualidade do Corte e Preço')
plt.xlabel('Qualidade do Corte')
plt.ylabel('Preço')
plt.show()
```

```python
# Cor (color) e preço
plt.figure(figsize=(10, 6))
sns.boxplot(x='color', y='price', data=df)
plt.title('Cor e Preço')
plt.xlabel('Cor')
plt.ylabel('Preço')
plt.show()
```

```python
# Pureza (clarity) e preço
plt.figure(figsize=(10, 6))
sns.boxplot(x='clarity', y='price', data=df)
plt.title('Pureza e Preço')
plt.xlabel('Pureza')
plt.ylabel('Preço')
plt.show()
```

```python
# Dimensões (x, y, z) e preço
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='price', data=df)
plt.title('Comprimento (x) e Preço')
plt.xlabel('Comprimento (x)')
plt.ylabel('Preço')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='y', y='price', data=df)
plt.title('Largura (y) e Preço')
plt.xlabel('Largura (y)')
plt.ylabel('Preço')
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(x='z', y='price', data=df)
plt.title('Profundidade (z) e Preço')
plt.xlabel('Profundidade (z)')
plt.ylabel('Preço')
plt.show()
```

## 2. Perguntas de partida e hipóteses

1. **Qual é a relação entre o preço do diamante e seu peso (carat)?**
   - Espera-se que diamantes com maior peso (carat) tenham preços mais elevados.

2. **Como a qualidade do corte (cut) influencia o preço?**
   - Diamantes com cortes de qualidade superior (ex: Ideal) serão mais caros.

3. **Existem cores (color) que são mais caras que outras?**
   - Diamantes de cor mais rara (ex: D) tendem a ser mais caros.

4. **O grau de pureza (clarity) tem um impacto significativo no preço?**
   - Diamantes com maior pureza (clarity) terão preços mais altos.

5. **Existem padrões nas dimensões (x, y, z) que possam prever o preço?**
   - O preço pode ser previsto a partir das dimensões do diamante.


## 3. Insights

1. Há uma correlação positiva entre o peso (carat) do diamante e seu preço.
2. Diamantes com cortes de qualidade superior tendem a ter preços mais altos, mas há variabilidade.
3. Diamantes de cores mais raras (ex: D, E, F) são geralmente mais caros.
4. Maior pureza (clarity) geralmente leva a preços mais altos, mas com alguma sobreposição entre as categorias.
5. As dimensões do diamante (x, y, z) mostram correlação com o preço, sugerindo que maiores tamanhos levam a preços mais altos.
